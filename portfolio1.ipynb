```python
# import libraries
import pandas as pd
import seaborn as sns
```

```python
# import dataset 
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
# Quick check on how the dataset looks like 
df.head() 
```

```python
# Check rows and columns
df.shape
```

```python
# Very important as we need to look for the column data types
df.dtypes
```

```python
# Clean up data 
# Only want USA races, 50km or 50mi 
# 2020
```

```python
df[df["Event distance/length"] == "50mi"]
```

```python
# combine 50km/50mi with isin
# We need to use "()" with "&"
df[(df["Event distance/length"].isin(["50km", "50mi"]))\
& (df["Year of event"] == 2020)]
```

```python
# only want USA races
df[df["Event name"].str.split("(").str.get(1).\
   str.split(")").str.get(0) == "USA"]
```

```python
# combine all the filters together
df2 = df[(df["Event distance/length"].isin(["50km", "50mi"]))\
& (df["Year of event"] == 2020)\
& (df["Event name"].str.split("(").str.get(1).\
   str.split(")").str.get(0) == "USA")]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
# remove (USA) from Event name
df2["Event name"] = df2["Event name"].str.split("(").str.get(0)
```

```python
# clean up athlete age
df2["athlete_age"] = 2020 - df2["Athlete year of birth"]
```

```python
df2.head(2)
```

```python
# remove h from Athlete performance
df2["Athlete performance"] = df2["Athlete performance"]\
.str.split(" ").str.get(0)
```

```python
df2.head(2)
```

```python
# drop columns: Athlete club, Athlete country, Athlete year of birth, 
# Athlete age category
df2 = df2.drop(["Athlete club", "Athlete country", "Athlete year of birth",\
    "Athlete age category"], axis=1)
    # need to include "axis=1" to specify columns 
```

```python
df2.head(1)
```

```python
# clean up null values
df2.isnull().sum()
```

```python
# Checking for data with null values, will output a dataframe  
df2[df2["athlete_age"].isnull() == 1]
# df2[df2["athlete_age"].isnull()]
```

```python
# drop null values
df2=df2.dropna()
```

```python
# Counting the numbe of duplicated rows
df2.duplicated().sum()
```

```python
# reset index
df2.reset_index(drop = True)
```

```python
# fix types
df2.dtypes
```

```python
df2["athlete_age"] = df2["athlete_age"].astype(int)
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
df2.dtypes
```

```python
# rename columns
df2 = df2.rename(columns={
    "Year of event":"year",
    "Event dates":"race_day",
    "Event name":"race_name",
    "Event distance/length":"race_length",
    "Event number of finishers":"race_number_of_finishers",
    "Athlete performance":"athlete_performance",
    "Athlete gender":"athlete_gender",
    "Athlete average speed":"athlete_average_speed",
    "Athlete ID":"athlete_id"
}
)
```

```python
df2.head(1)
```

```python
# reorder columns
df3 =df2[["race_day", "race_name", "race_length", "race_number_of_finishers", "athlete_id", "athlete_gender", "athlete_age", "athlete_performance", "athlete_average_speed"]]
```

```python
df3.head(2)
```

```python
# charts & graphs
sns.histplot(df3, x="race_length", hue="athlete_gender")
```

```python
sns.distplot(df3[df3["race_length"] == "50mi"]["athlete_average_speed"])
```

```python
sns.violinplot(data=df3, x="race_length", y="athlete_average_speed",\
    hue="athlete_gender", split=True, inner="quart", linewidth=1)
```

```python
# lmplot: scatter plot & regression line
sns.lmplot(data=df3, x="athlete_age", y="athlete_average_speed",\
    hue="athlete_gender")
```

```python
# questions I want to find out from the dataset
# race_day
# race_name
# race_length
# race_number_of_finishers
# athlete_performance
# athlete_gender
# athlete_average_speed
# athlete_id
```

```python
# difference in speed for the 50km/50mi male to female
df3.groupby(["race_length", "athlete_gender"])["athlete_average_speed"]\
    .mean()
```

```python
# what age groups are the best in 50mi Race (20 + races min)
df3[df3["race_length"] == "50mi"].groupby("athlete_age")["athlete_average_speed"].\
    agg(["mean", "count"]).sort_values("mean", ascending=False).query("count>19")
```

```python
# Seasons for the data -> Slower in summer than winter?
# spring: 3-5
# summer: 6-8
# fall: 9-11
# winter: 12-2
```

```python
df3["race_month"] = df3["race_day"].str.split(".").str.get(1).astype(int)
```

```python
df3.head(2)
```

```python
df3["race_season"] = df3["race_month"].apply(lambda x: "Winter" if x > 11 else "Fall" if x > 8 else "Summer" if x > 5 else "Spring")
```

```python
df3.head(2)
```

```python
df3.groupby("race_season")["athlete_average_speed"].agg(["mean", "count"]).\
    sort_values("mean", ascending=False)
```

```python
# only 50 milers
df3.query('race_length == "50mi"').groupby("race_season")["athlete_average_speed"].\
    agg(["mean", "count"]).sort_values("mean", ascending=False)
```
